# Launching Label Studio from Jupyter

This notebook launches a local Label Studio server.

📌 **Tip**: If multiple users work simultaneously, you should change the default port (8080) by modifying the `YOURPORT` variable in the first cell.

🔗 See Label Studio official documentation for more:  
https://labelstud.io/guide/start.html#Run-Label-Studio-on-localhost-with-a-different-port

**Warning**
Before launching this notebook, make sure that the kernel you have chosen contains the necessary dependencies to launch Label Studio.

```python
#Insert your jupyter name (jupyter-name)

PATHTOYOURENVIRONMENT = ""
YOURPORT = "8080"
```

```python
!bash -c "source $PATHTOYOURENVIRONMENT/bin/activate && \
export LABEL_STUDIO_PORT=$YOURPORT && \
export LABEL_STUDIO_HOST=http://172.16.100.15:$YOURPORT && \
export LABEL_STUDIO_LOCAL_FILES_SERVING_ENABLED=true && \
export LOCAL_FILES_DOCUMENT_ROOT=/ && \
label-studio start"
```
